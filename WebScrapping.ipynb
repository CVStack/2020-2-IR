```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('http://web.kma.go.kr/home/index.tab.now-ten.jsp?gubun=1&myPointCode=&unit=M') #get resuource 
soup = BeautifulSoup(page.content, 'html.parser')

today_clock = soup.find('p', {'class' : 'clock'}).get_text()
soup = soup.find('div', {'class' : 'weather'})

regions = soup.find_all('dl')

region_info_dict = {}
print(today_clock)
for region in regions:
    region_name = region.find('dt').get_text()
    region_weather = region.find('dd').find('img')['alt']
    region_temp = region.find('dd', {'class' : 'temp'}).get_text()
    region_info_dict[region_name] = {'weather' : region_weather, 'temp' : region_temp}
    
    
    
page = requests.get('http://web.kma.go.kr/home/index.tab.now-ten.jsp?gubun=2&myPointCode=&unit=M') #get resuource 
soup = BeautifulSoup(page.content, 'html.parser')
soup = soup.find('div', {'class' : 'rainfall'})
regions = soup.find_all('dl')

for region in regions:
    region_name = region.find('dt').get_text()
    region_rainfall = region.find('dd')
    if region_rainfall is None :
        region_rainfall = '0.0mm'
    else :
        region_rainfall = region_rainfall.get_text()
        
    region_info_dict[region_name]['rainfall'] = region_rainfall
    
region_info_dict
```
