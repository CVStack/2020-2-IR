```
import matplotlib.pyplot as plt
import numpy as np
points = [
          [0.3, 0.3],
          [0.4, 0.4],
          [0.7, 0.35],
          [0.35, 0.8],
          [0.6, 0.6],
          [0.9, 0.8],
          [0.6, 0.1],
          [0.9, 0.05],
          [0.55, 0.9],
          [0.8, 0.55],
]
points = np.array(points)

plt.scatter(points[:, :1], points[:, 1:])
# plt.show()
window = []

for point in points:
  if len(window) == 0:
    window.append(point)
    continue
  
  t = 0
  for i in range(len(window)):
    element = window[i]

    if element[0] > point[0] and element[1] > point[1]:
      element[0] = point[0], element[1] = point[1]
      window[i] = element
      break
    elif element[0] < point[0] and element[1] < point[1]:
      break
    t += 1

  if t == len(window):
    window.append(point)

print(window)
window = np.array(window)
plt.scatter(window[:, :1], window[:, 1:], c='r')
plt.show()
```
