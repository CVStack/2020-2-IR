```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

example_sents = [
    "breakthrough drug for schizophrenia", #doc0
    "new schizophrenia drug", #doc1
    "new approach for treatment of schizophrenia", #doc2
    "new hopes for schizophrenia patients" #doc3
]



def textTransformation(sents):
    rs_sents = []
    indexed_words = []
    for example_sent in sents:
        stop_words = set(stopwords.words('english'))
        word_tokens = word_tokenize(example_sent) #tokenization
        filtered_sentence = []

        for w in word_tokens: #removal stopwords
            if w not in stop_words:
                filtered_sentence.append(w)

#         print(filtered_sentence)

        from nltk.stem import PorterStemmer

        ps = PorterStemmer()
        sent = []
        for w in filtered_sentence: #stemming
#             print(w, " : ", ps.stem(w))
            sent.append(ps.stem(w))
            indexed_words.append(ps.stem(w))
        rs_sents.append(sent)
#         print()
    return rs_sents, list(set(indexed_words))
```

```python
rs_sents, indexed_words = textTransformation(example_sents)
# print(indexed_words)
# print(rs_sents)
vectors = []

for sent in rs_sents:
    vector = [0] * len(indexed_words)
    i = 0
    for i in range(len(indexed_words)):
        if indexed_words[i] in sent:
            vector[i] = 1
    vectors.append(vector)


querys = [
    "schizophrenia drug",
    "schizophrenia patient"
]

print()
querys = textTransformation(querys)[0]
query_vector = []
for query in querys:
    vector = [0] * len(indexed_words)
    i = 0
    for i in range(len(indexed_words)):
        if indexed_words[i] in query:
            vector[i] = 1
    query_vector.append(vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

for qv in query_vector:
    scores = []
    for v in vectors:
        rs = cosine_similarity([qv], [v])
        scores.append(rs)
    print(scores)
    print("Best --> doc {} : {}".format(scores.index(max(scores)), max(scores)))
    print()
```
